# Amostragem Sistematica em um CSV Randon

## a amostragem sistemática é um método estatístico que envolve a seleção de elementos de uma base de amostragem ordenada.

- População 4860,
- Amostra 16,
- Intervalos 300,


```python
import pandas as pd
import numpy as np
import random 
```

```python
dados = pd.read_csv("Korean_demographics_2000-2022.csv")
dados.shape
```

```python
def AmostragemSistematica(dados, amostras):
    intervalo = len(dados) // amostras
    random.seed(1)
    start = random.randint(0,intervalo)
    indice = np.arange(start, len(dados), step= 300)
    Amostragem_Sistematica = dados.iloc[indice]
    return Amostragem_Sistematica
```

```python
Dataf_Amostra_Sistematica = AmostragemSistematica(dados,16)

Dataf_Amostra_Sistematica

```
